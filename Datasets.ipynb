We will utilize numerical attributes  $21$ to $39$ to construct matrix $X$. In column $40$, we have the class label, which will be used to define vector $y$. Here is a summary of these variables.

Cleaning data

```python
df = df.iloc[:, 20:40]
df = df.dropna()
df = df.reset_index(drop=True)
#Remove '?' from the dataset
df = df[(df != '?').all(axis=1)]
df = df.reset_index(drop=True)

```

```python
x = df.iloc[:, 0:19].values
y = df.iloc[:, -1].values


y, _ = pd.factorize(y)
y = y.reshape(-1, 1)

print(y)
savemat(path + '/python/bands.mat', {'x': x, 'y': y})
dfNew = pd.DataFrame(np.concatenate((x, y), axis=1))
dfNew.to_csv(path + '/python/bands.csv', index=False, header=False)
```

# Contraceptive Method Choice
http://archive.ics.uci.edu/dataset/30/contraceptive+method+choice

```python
df = pd.read_csv(path + 'contraceptive+method+choice/cmc.data.txt', header=None)
df = df.astype('category')
df.head()
```

| Attribute | Description | Domain |
|-----------|-------------|--------|
| Wife's age | numerical |  |
| Wife's education | categorical | 1=low, 2, 3, 4=high |
| Husband's education | categorical | 1=low, 2, 3, 4=high |
| Number of children ever born | numerical |  |
| Wife's religion | binary | 0=Non-Islam, 1=Islam |
| Wife's now working? | binary | 0=Yes, 1=No |
| Husband's occupation | categorical | 1, 2, 3, 4 |
| Standard-of-living index | categorical | 1=low, 2, 3, 4=high |
| Media exposure | binary | 0=Good, 1=Not good |
| Contraceptive method used | class attribute | 1=No-use 2=Long-term 3=Short-term |


Cleaning data   

```python
df = df.iloc[:, 0:10]
df = df.dropna()
#Remove '?' from the dataset
df = df[(df != '?').all(axis=1)]
df = df.reset_index(drop=True)
```

Splitting data into $X$ and $y$

```python
x = df.iloc[:, 0:9].values
y = df.iloc[:, -1].values
print(y)
```

Convert the categorical $y$ data in  into numerical. The equivalence is as folows.

    1.Keep the categorical 1
    2.The categoricals 2 and 3 are converted to 2

```python
for i in range(len(y)):
    if y[i] == 1:
        y[i] = 1
    else:
        y[i] = 2
  
y, _ = pd.factorize(y)
y = y.reshape(-1, 1)
print(y)

savemat(path + '/python/cmc.mat', {'x': x, 'y': y})
dfNew = pd.DataFrame(np.concatenate((x, y), axis=1))
dfNew.to_csv(path + '/python/cmc.csv', index=False, header=False)
```

# Sonar 

http://archive.ics.uci.edu/dataset/151/connectionist+bench+sonar+mines+vs+rocks

```python
df = pd.read_csv(path + 'sonar/sonar.all-data.csv', header=None)
df = df.astype('category')
df.head()
```

Features are all continuous and are all about frequency response of sonar chirp returns bouncing off different surfaces. The label associated with each record contains the letter R if the object is a rock and M if it is a mine (metal cylinder).

```python
#Remover a primeira linha
df = df.iloc[1:, :]
df = df.iloc[:, 0:61]
df = df.dropna()
#Remove '?' from the dataset
df = df[(df != '?').all(axis=1)]
df = df.reset_index(drop=True)
print(df.head())
```

Splitting data into $X$ and $y$

Y follows the following equivalence:

    1.Keep the categorical R
    2.The categoricals M are converted to 2

```python
X = df.iloc[:, 0:60].values
y = df.iloc[:, -1].values
print(y)
yNew = np.zeros((len(y), 1))
for i in range(len(y)):
    if y[i] == 'R':
        yNew[i] = 1
    else:
        yNew[i] = 2


#Converter para int
yNew = yNew.astype(int)
print(yNew)
```

```python

savemat(path + '/python/sonar.mat', {'x': X, 'y': yNew})
dfNew = pd.DataFrame(np.concatenate((X, yNew), axis=1))
dfNew.to_csv(path + '/python/sonar.csv', index=False, header=False)
```
